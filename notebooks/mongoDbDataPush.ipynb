```python
# Convert the dataframe into dict
data_dict = df.to_dict(orient = "records")
data_dict
```

## steps to push the data into atlas collection
---
1. Create the database and cluster(create from atlas website)
2. Here create a database
3. Inside the database create a collection
4. push data into the collection

```python
import os
CONNECTION_URL = os.getenv("mongo_connection_url")
DATABASE_NAME = "laptop_price"
COLLECTION_NAME = "laptop_price_collection"
```

```python
# create the client for mongoDb
client = MongoClient(CONNECTION_URL)
```

```python
# create the database
database = client[DATABASE_NAME]
# create the collection
collection = database[COLLECTION_NAME]
```

```python
# now insert the data into the collection
records = collection.insert_many(data_dict)
```

```python
records
```

```python
# get all the data from database
all_data = collection.find()
```
